```python
# ðŸ“˜ Step 1: Import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ðŸ“˜ Step 2: Define the function and its derivative

# Function: y = (x + 3)^2
def f(x):
    return (x + 3)**2

# Derivative (Gradient): f'(x) = 2(x + 3)
def df(x):
    return 2 * (x + 3)
```

```python
# ðŸ“˜ Step 3: Initialize parameters

x = 2               # starting point (given)
learning_rate = 0.1 # step size (controls how fast we move)
epochs = 50         # number of iterations
tolerance = 1e-6    # stop when improvement is very small

# Store progress for visualization
x_history = [x]
y_history = [f(x)]
```

```python
# ðŸ“˜ Step 4: Implement Gradient Descent Loop

for i in range(epochs):
    gradient = df(x)                  # find slope at current x
    new_x = x - learning_rate * gradient  # move opposite to slope

    # Stop if improvement is very small
    if abs(new_x - x) < tolerance:
        print(f"Converged at iteration {i}")
        break

    x = new_x
    x_history.append(x)
    y_history.append(f(x))

# Final result
print(f"âœ… Local minimum occurs at x = {x:.4f}")
print(f"Function value at minimum (y) = {f(x):.4f}")
```

```python
# ðŸ“˜ Step 5: Visualize the Gradient Descent Process

X = np.linspace(-6, 4, 100)
Y = f(X)

plt.figure(figsize=(10,6))
plt.plot(X, Y, label="y = (x + 3)Â²")
plt.scatter(x_history, y_history, color='red', label='Descent steps')
plt.plot(x_history, y_history, color='red', linestyle='--')
plt.title("Gradient Descent Convergence to Local Minimum")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
